# Importing Packages

```python
# Import required packages
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm
```

# Loading Data

```python
# Use numpy.loadtxt to import our data
omega, ph_tot, ph_c1, ph_c2 = np.loadtxt('gr.dos', unpack=True)
```

# Plotting Our Data

```python
# Create figure and add axes object
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# Plot our data
ax.plot(omega, ph_tot)
ax.plot(omega, ph_c1)

# Save and show our data
plt.savefig('plot.pdf', bbox_inches='tight')
plt.show()
```

### Removing Spines

# Final Edited Plot

### Plotting Script

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 22
plt.rcParams['axes.linewidth'] = 2

# Use numpy.loadtxt to import our data
omega, ph_tot, ph_c1, ph_c2 = np.loadtxt('gr.dos', unpack=True)

# Create figure object and store it in a variable called 'fig'
fig = plt.figure()

# Add axes object to our figure that takes up entire figure
ax = fig.add_axes([0, 0, 1, 1])

# Add the x and y-axis labels
ax.set_xlabel(r'Frequency (cm$^{-1}$)')
ax.set_ylabel(r'Phonon DOS (state/cm$^{-1}$/unit-cell)')

# Edit the major and minor ticks of the x and y axes
ax.xaxis.set_tick_params(which='major', size=12, width=2, direction='in', right='on')
ax.xaxis.set_tick_params(which='minor', size=6, width=2, direction='in', right='on')
ax.yaxis.set_tick_params(which='major', size=12, width=2, direction='in', right='on')
ax.yaxis.set_tick_params(which='minor', size=6, width=2, direction='in', right='on')

# Edit the major and minor tick locations of x and y axes
ax.xaxis.set_major_locator(mpl.ticker.MultipleLocator(500))
ax.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(100))
ax.yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.01))
ax.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.002))

# Set the axis limits
ax.set_xlim(0, 1700)
ax.set_ylim(0, 0.025)

# Plot our data
ax.plot(omega, ph_tot, linewidth=2, color='blue', linestyle='solid')
ax.plot(omega, ph_c1, linewidth=2, color='red', linestyle='dashed')

# Add annotate
ax.annotate('Total', color='blue',
            xy=(1390, 0.015), xycoords='data',
            xytext=(1000, 0.018), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='blue')
           )
ax.annotate('C atom', color='red',
            xy=(1300, 0.002), xycoords='data',
            xytext=(950, 0.01), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='red')
           )

# Save and show figure
plt.savefig('plot-rev.pdf', bbox_inches='tight')
plt.show()
```
